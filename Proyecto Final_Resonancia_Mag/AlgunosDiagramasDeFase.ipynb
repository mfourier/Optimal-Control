```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Elegir carpeta de solucion
str_fold = "exportalfa3"
str_fold_bocop = "exportContrastReg"
```

```python
#Se cargan los datos
times = pd.read_table(str_fold+'/discretization_times.export').to_numpy()
y_1 = pd.read_table(str_fold+'/y1.export').to_numpy()
z_1 = pd.read_table(str_fold+'/z1.export').to_numpy()
y_2 = pd.read_table(str_fold+'/y2.export').to_numpy()
z_2 = pd.read_table(str_fold+'/z2.export').to_numpy()
u = pd.read_table(str_fold+'/ux.export').to_numpy()
```

```python
#Variables para dibujar el circulo unitario
# y = -+sqrt(1-x^2)
x_vec = np.linspace(-1,1,1000)
y_vec_1 = np.sqrt(1-np.power(x_vec,2))
y_vec_2 = -np.sqrt(1-np.power(x_vec,2))

#plt.plot(x_vec,y_vec_1, linestyle ='dashed', color='black')
#plt.plot(x_vec,y_vec_2, linestyle ='dashed', color='black')
```

```python
#Diagrama de fase q1
fig = plt.figure(figsize = (8,6)) 
plt.plot(y_1,z_1, color = 'purple')
plt.ylabel('$z_1$')
plt.xlabel('$y_1$')
plt.scatter(y_1[0],z_1[0], label = 'Starting Point', color = 'red')
plt.legend()
plt.title("Diagrama de fases para $q_1 = (y_1,z_1)$")
plt.plot(x_vec,y_vec_1, linestyle ='dashed', color='black')
plt.plot(x_vec,y_vec_2, linestyle ='dashed', color='black')
plt.show()
```

```python
#Diagrama de fase q1 q2
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=((12,6)))
fig.suptitle('Diagrama de fases \n $P_1$: Fluido')
ax1.plot(y_1,z_1, color = 'purple')
ax2.plot(y_2,z_2, color = 'purple')
ax1.plot(x_vec,y_vec_1, linestyle ='dashed', color='black')
ax1.plot(x_vec,y_vec_2, linestyle ='dashed', color='black')
ax2.plot(x_vec,y_vec_1, linestyle ='dashed', color='black')
ax2.plot(x_vec,y_vec_2, linestyle ='dashed', color='black')
ax1.set_title('$q_1$')
ax2.set_title('$q_2$')

ax1.set_xlabel("$y_1$")
ax1.set_ylabel("$z_1$")
ax2.set_xlabel("$y_2$")
ax2.set_ylabel("$z_2$")
plt.show()
```

```python
#Grafico del control
fig = plt.figure(figsize = (6,4)) 
plt.plot(times,u, color = 'black')
plt.title("u")
plt.xlabel('t')
plt.show()
```
